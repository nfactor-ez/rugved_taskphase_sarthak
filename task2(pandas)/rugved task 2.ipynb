```python
import pandas as pd
df =pd.read_csv(r'C:\Users\sarth\OneDrive\Desktop\matches.csv')


year_2008_matches = df[df['season'] == 2008].shape[0]

print('Total number of matches conducted in 2008:', year_2008_matches)
```

```python
city_match_counts = df['city'].value_counts()
max_city = city_match_counts.index[1]
min_city = city_match_counts.idxmin()
max_count = city_match_counts.max()
min_count = city_match_counts.min()

print('City with maximum matches:', max_city, 'with', max_count, 'matches')
print('City with minimum matches:', min_city, 'with', min_count, 'matches')
```

```python
city_wise_counts = df['city'].value_counts()

print('City-wise match counts:')
print(city_wise_counts)
```

```python
toss_counts = df['toss_winner'].value_counts()
print('Team with most toss wins:', toss_counts.index[0], 'with', toss_counts.iloc[0], 'wins')
print('Team with least toss wins:', toss_counts.index[-1], 'with', toss_counts.iloc[-1], 'wins')

```

```python
toss_decisions = df['toss_decision'].value_counts()
print('Toss decisions breakdown:')
print(toss_decisions)
```

```python
result_counts = df['result'].value_counts()
print('Match result types:')
print(result_counts)
```

```python
tie_matches = df[df['result'] == 'tie']
print('Teams involved in tie matches:')
if not tie_matches.empty:
    print('Team 1:', tie_matches['team1'].tolist())
    print('Team 2:', tie_matches['team2'].tolist())
else:
    print('No tie matches found')
```

```python
highest_run_win = df.loc[df['win_by_runs'].idxmax()]
print('Highest run margin victory:')
print('Winner:', highest_run_win['winner'])
print('Margin:', highest_run_win['win_by_runs'], 'runs')
```

```python
lowest_run_win = df.loc[df['win_by_runs'].idxmin()]
print('Highest run margin victory:')
print('Winner:', lowest_run_win['winner'])
print('Margin:', lowest_run_win['win_by_runs'], 'runs')
```

```python
pom_counts = df['player_of_match'].value_counts()
players_more_than_3 = pom_counts[pom_counts > 3]

print("Players with more than 3 Player of the Match awards:")
print(players_more_than_3)
```

```python
highest_runs = df.loc[df['win_by_runs'].idxmax()]
lowest_runs = df.loc[df['win_by_runs'][df['win_by_runs'] > 0].idxmin()]
print("Venue with highest run margin win:", highest_runs['venue'], "(" + str(highest_runs['win_by_runs']), "runs)")
print("Venue with lowest run margin win:", lowest_runs['venue'], "(" + str(lowest_runs['win_by_runs']), "runs)")
```

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
df =pd.read_csv(r'C:\Users\sarth\OneDrive\Desktop\matches.csv')

deliveries_df = pd.read_csv(r'C:\Users\sarth\OneDrive\Desktop\deliveries.csv')
matches_per_season = df.groupby('season').size()
print(matches_per_season)
```

```python

merged_data = pd.merge(deliveries_df, df, left_on='match_id', right_on='id')

runs_per_season = merged_data.groupby('season')['total_runs'].sum()
print(runs_per_season)
```

```python
toss_decision_per_season = df.groupby(['season', 'toss_decision']).size().unstack()

# Plotting
toss_decision_per_season.plot(kind='bar', figsize=(10, 5))
plt.title("Toss Decisions Across Seasons")
plt.ylabel("Number of Toss Decisions")
plt.xlabel("Season")
plt.show()
```

```python
dismissals = deliveries_df['dismissal_kind'].value_counts()
dismissals.plot(kind='bar',figsize=(20,6))
plt.title("DISMISSAL KINDS & COUNT")
plt.xlabel("count")
plt.ylabel("dismissal kind")
plt.show()
```

```python
top_scorers = deliveries_df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)


top_scorers.plot(kind='bar', figsize=(10, 6))
plt.title("Top 10 Run Scorers in IPL")
plt.ylabel("Total Runs")
plt.xlabel("Batsman")
plt.show()
```

```python
mom_awards=df['player_of_match'].value_counts().head(10)
mom_awards.plot(kind='bar',figsize=(20,6))
plt.title("top 10 players with mom awards")
plt.ylabel("Number of Awards")
plt.xlabel("Player")
plt.show()
```
