```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
penguin_df = pd.read_csv(r'C:\Users\sarth\OneDrive\Desktop\penguins.csv')
penguin_df.head()
```

```python

plt.scatter(penguin_df['body_mass_g'],penguin_df['culmen_depth_mm'])
```

```python
plt.scatter(penguin_df['culmen_length_mm'],penguin_df['culmen_depth_mm'])
```

```python
plt.scatter(penguin_df['culmen_length_mm'],penguin_df['body_mass_g'])
```

```python
penguins_cleaned = penguin_df.dropna(subset=['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'])
feat= penguins_cleaned[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]

```

```python
scaler=StandardScaler()

features=scaler.fit_transform(feat)
```

```python
inertia = []
max_cluster=10
for n_clusters in range(1, max_cluster + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, max_cluster + 1), inertia, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, max_cluster + 1))
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
penguins_cleaned['cluster'] = kmeans.fit_predict(features)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins_cleaned, x='body_mass_g', y='culmen_depth_mm', hue='cluster', palette='viridis', s=100)
plt.title('K-Means Clustering of Penguins')
plt.xlabel('body_mass')
plt.ylabel('Culmen Depth (mm)')
plt.legend(title='Cluster')
plt.show()

print('K-Means clustering implemented and visualized.')
```
